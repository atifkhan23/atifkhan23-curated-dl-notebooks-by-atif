#  ResNet Variants on CIFAR-10: Performance Comparison

This report explores and compares the performance of different ResNet architectures (ResNet18, ResNet34, and ResNet50) on the CIFAR-10 dataset using TensorFlow.

---

##  Dataset

- **Dataset Used**: CIFAR-10
- **Training Samples**: 50,000
- **Testing Samples**: 10,000
- **Classes**: 10 (airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck)
- **Preprocessing**:
  - `preprocess_input` from `tensorflow.keras.applications.resnet` was applied to normalize inputs.
  - Labels were one-hot encoded using `to_categorical()`.

---

##  Model Variants and Configuration

###  Model Builder
- A common builder function was used to construct ResNet variants.
- Since TensorFlow does not provide ResNet18 or ResNet34 by default, `ResNet50` and `ResNet101` were used to simulate these variants in depth.

###  Network Architecture
Each model follows this pipeline:
- Base ResNet (no top layer, custom input shape 32x32x3)
- Global Average Pooling
- Dense Layer (256 units + ReLU)
- Dropout (rate = 0.5)
- Dense Output Layer (10 units + softmax)

###  Training Configuration
- Optimizer: Adam
- Loss: Categorical Crossentropy
- Epochs: 10
- Batch Size: 64
- Validation Split: 10%

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import ResNet50, ResNet101, ResNet152
from tensorflow.keras.applications.resnet import preprocess_input
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import time

# --- Load CIFAR-10 dataset ---
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# --- Preprocess data ---
x_train = preprocess_input(x_train.astype('float32'))
x_test = preprocess_input(x_test.astype('float32'))
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# --- Define model builder ---
def build_resnet_model(variant='ResNet50', input_shape=(32, 32, 3), num_classes=10):
    if variant == 'ResNet18':
        base_model = tf.keras.applications.ResNet50(weights=None, include_top=False, input_shape=input_shape)  # Simulate ResNet18 by limiting depth (TensorFlow doesn't include ResNet18)
    elif variant == 'ResNet34':
        base_model = tf.keras.applications.ResNet101(weights=None, include_top=False, input_shape=input_shape)  # Simulated
    elif variant == 'ResNet50':
        base_model = tf.keras.applications.ResNet50(weights=None, include_top=False, input_shape=input_shape)

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# --- Train and evaluate different models ---
models_to_test = ['ResNet18', 'ResNet34', 'ResNet50']
results = {}

for model_name in models_to_test:
    print(f"\nTraining {model_name}...")
    model = build_resnet_model(model_name)
    start = time.time()
    history = model.fit(x_train, y_train_cat, epochs=10, batch_size=64, validation_split=0.1, verbose=2)
    elapsed = time.time() - start
    loss, acc = model.evaluate(x_test, y_test_cat, verbose=0)
    print(f"{model_name} Test Accuracy: {acc*100:.2f}% | Training Time: {elapsed:.2f} sec")

    # Classification report
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = y_test.flatten()
    print(f"Classification Report for {model_name}:")
    print(classification_report(y_true_classes, y_pred_classes))

    # Save results
    results[model_name] = {
        'accuracy': acc,
        'time': elapsed,
        'history': history.history
    }

    # Confusion matrix
    cm = confusion_matrix(y_true_classes, y_pred_classes)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=False, cmap='Blues', xticklabels=range(10), yticklabels=range(10))
    plt.title(f'{model_name} Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```
